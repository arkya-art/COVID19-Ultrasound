```python
import pandas as pd
import os
import shutil
```

```python
FILE_PATH = "chestxray/metadata(1).csv"
IMAGES_PATH = "chestxray/images"
```

```python
df = pd.read_csv(FILE_PATH)
df.head()
```

```python
df.shape
```

```python
TARGET_DIR = "Dataset/Covid"

if not os.path.exists(TARGET_DIR): # if target path not exist then create it
    
    os.mkdir(TARGET_DIR)
    print("Covid folder created")
    
```

```python
# taking out all the images from covid dataset and place it inside new folder Covid

cnt = 0

for (i,row) in df.iterrows(): # iterating thru the rows and selecting front view covid chest xray
    
    if row["finding"] == "COVID-19" and row["view"] == "PA":
        
        # copy all the selected images from old directory to new directory
        filename = row['filename']
        image_path = os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        print("Moving image",cnt)
        cnt += 1
        
print(cnt) 

# there exist 142 images of covid19 with front view x ray
```

```python
# sampling of images from kaggle

import random
KAGGLE_FILE_PATH = 'chest_xray_kaggle/train/NORMAL'
TARGET_NORMAL_DIR = "Dataset/Normal"
```

```python
image_names = os.listdir(KAGGLE_FILE_PATH)
image_names # lists all the images in the normal folder
```

```python
random.shuffle(image_names)
```

```python
# picking 1st 142 images from randomly shuffled normal images
for i in range(142):
    
    image_name = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH,image_name)
    
    target_path = os.path.join(TARGET_NORMAL_DIR,image_name)
    
    shutil.copy2(image_path,target_path)
    print("copying image", i)
```
